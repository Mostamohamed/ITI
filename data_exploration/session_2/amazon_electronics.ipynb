```python
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301')
```

```python
import urllib
df=pd.DataFrame([],columns=['title','price','url','total_rating'])
products_Sections = driver.find_elements(By.XPATH,"//div[contains(@class, 's-main-slot')]//div[@data-component-type='s-search-result']")
for section in products_Sections:
    try:
        title=section.find_element(By.XPATH,".//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']/span")
        print(title.text)
        price=section.find_element(By.XPATH,".//span[@class='a-price-whole']")
        print(price.text)
        image_src=section.find_element(By.XPATH,".//img[@class='s-image']")
        url=image_src.get_attribute('src')
        print(url)
        total_rating=section.find_element(By.XPATH,".//span[@class='a-size-base s-underline-text']")
        print(total_rating.text)
        df.loc[len(df)]=[title.text,price.text,url,total_rating.text]
    except:
        print('error')
```
