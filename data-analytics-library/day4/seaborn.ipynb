```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(x='year',y='passengers',data=flights)

plt.title('passengers over time')
plt.xlabel('year')
plt.ylabel('passengers')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x='total_bill',y='tip',data=tips,hue='time',palette='Set1')

plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
plt.figure(figsize=(8,6))

sns.barplot(x='day',y='total_bill',errorbar=None,data=tips,hue='sex',palette='Set1')

plt.title('Average Total Bill by Day')
plt.xlabel('day')
plt.ylabel('Average Total Bill ')
```

```python
plt.figure(figsize=(8,6))

sns.histplot(x=tips['total_bill'],bins=10)

plt.title('dist of Total Bill')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.show()
```
